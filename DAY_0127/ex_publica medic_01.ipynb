## 시도 이름을 표준 행정구역 이름으로 변경

```python
#경기=>경기도, 경남=> 경상남도 등
```

```python
## 표준 행정구역 이름으로 수정:#경기=>경기도, 경남=> 경상남도 등

# addr_aliases={'경기':'경기도','경남':'경상남도','경북':'경상북도','충북':'충청북도',
#              '서울시':'서울특별시','부산시':'부산광역시','대전시':'대전광역시'}
```

```python
addr['시도'].unique()
```

```python
## 표준 행정구역 이름으로 수정:#경기=>경기도, 경남=> 경상남도 등
addr_aliases = {'경기':'경기도', '경남':'경상남도', '경북':'경상북도', '충북':'충청북도',
'서울시': '서울특별시', '부산특별시': '부산광역시', '대전시': '대전광역시', '충남': '충청남도',
'전남': '전라남도', '전부': '전라북도'}

#dict.get(key[,default])
#-key에 해당하는 값이 없으면 defaulot값(key)을 리턴
addr['시도']=addr['시도'].apply(lambda v:addr_aliases.get(v,v))
```

## (2)'시도군구'를 기준으로 그룹을 만들고 그룹별 의료기관 수 합계구하기
- ['시도'],['군구'],['시도군구']를 기준으로 그룹 생성 후 그룹별 의료 기관수 합계 구하기



```python
addr_group = pd.DataFrame(addr.groupby(['시도', '군구', '시도군구'], as_index=False).count())
```

```python
addr_group = addr_group.set_index('시도군구')

addr_group. head()
```

```python
population=pd.read_excel('행정구역시군구별성별인구수.xlsx',sheet_name='데이터',header=[0])
population.head()
```

```python
# 행정구역 이름 변경
```

```python
population=population.rename(columns= {'행정구역(시도)별1':'시도',
                                      '행정구역(시군구)별2':'군구'})
population.head()
```

```python
#병합에 사용할 공통 컬럼 추가
```

```python
#['군구']컬럼에서 공백 제거:앞뒤 공백이 존재
```

```python
row_count=population.shape[0] #row 의 개수

for index in range(0,row_count):
    population['군구'][index]=population['군구'][index].strip()
    
```

```python
population['시도군구']=population.apply(lambda r:r['시도']+' '+r['군구'],axis=1)

population.head()
```

```python
#'군구'컬럼에서 '소계' 항목 제거
```

```python
population=population[population.군구 !='소계']  #특정 컬럼 접근 방법1
#population=population[population['군구']!='소계'] #특정 컬럼 접근 방법2
population.head()
```

```python
population=population.set_index('시도군구')
population.head()
```

```python
addr_population_merge = pd.merge(addr_group, population, how='inner',
left_index=True, right_index=True)

addr_population_merge.head()
```

```python
#필요한 컬럼만 추출
local_medicalcenter_population=addr_population_merge[['시도_x','군구_x','의료기관수','총인구수 (명)']]
```

```python
# 컬럼이름 변경
```

```python
local_medicalcenter_population=\local_medicalcenter_population.rename(columns={'시도_x':'시도','군구_x':'군구','의료기관수':'공공의료기관수','총인구수(명)':'인구수'})

local_medicalcenter_population.head()
```

```python
mc_count=local_medicalcenter_population['공공의료기관수']

local_medicalcenter_population['인구대비의료기관비율']= mc_count.div(local_medicalcenter_population['인구수'],axis=0)*100000

local_medicalcenter_population.head()
```

```python
import matplotlib.pyplot as plt
import platform
import koreanize_matplotlib

mc_ratio = local_medicalcenter_population[['공공의료기관수']]
mc_ratio = mc_ratio.sort_values(by='공공의료기관수', ascending=False)
mc_ratio.plot(kind='bar', rot=90, figsize=(30, 6))
```
